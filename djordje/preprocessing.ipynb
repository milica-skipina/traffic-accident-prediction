# WEATHER

### Scrape data from weather API
https://api.weather.com/v1/location/KLGA:9:US/observations/historical.json?apiKey=6532d6454b8aa370768e63d6ba5a832e&units=e&startDate=20110303

```python
import pytz
from datetime import datetime, date, timedelta
import requests
import json

def convert_timestamp_to_datetime_utc(timestamp):
    """
        Takes timestamp and converts it to utc time in US/Eastern timezone
    """
    utc_time = datetime.utcfromtimestamp(timestamp)
    return utc_time - timedelta(hours=4)
```

```python
BASE_URL = 'https://api.weather.com/v1/location/KLGA:9:US/observations/historical.json?apiKey=6532d6454b8aa370768e63d6ba5a832e&units=e&startDate='
start_date = date(2014, 1, 1)
end_date = date(2020, 1, 1)
delta = timedelta(days=1)

df = pd.DataFrame(columns=['Datetime', 'Temperature', 'FeelsLike', 'Dew', 'Humidity', 'Wind', 'WindSpeed', 'WindGust', 'Pressure', 'Precip.', 'Condition', 'Day/Night', 'Clouds'])

while start_date <= end_date:
    url = BASE_URL + start_date.strftime("%Y%m%d")
    print(start_date.strftime("%Y%m%d"))
    r = requests.get(url)
    if r.status_code != 200:
        print('ERROR')
    observations = r.json()['observations']
    
    result_dict = {convert_timestamp_to_datetime_utc(o['valid_time_gmt']).strftime("%Y-%m-%d %H:00"): o for o in observations}
    for key, value in result_dict.items():
        df = df.append({'Datetime': key, 'Temperature': value['temp'], 'FeelsLike': value['feels_like'], 'Dew': value['dewPt'], 
                        'Humidity': value['rh'], 'Wind': value['wdir_cardinal'], 'WindSpeed': value['wspd'], 'WindGust': value['gust'],
                        'Pressure': value['pressure'], 'Precip.': value['precip_hrly'], 'Condition': value['wx_phrase'], 
                        'Day/Night': value['day_ind'], 'Clouds': value['clds']}, ignore_index=True)
        
    start_date += delta
    
df.to_csv('data/weather.csv', index=False)
```

```python
#'LBoro', 'RBoro', 'BikeLane', 'BIKE_TRAFDIR'
geo_data2 = geo_data[['Street', 'NonPed', 'StreetWidth_Min', 'XFrom', 'YFrom', 'XTo',
                      'YTo', 'POSTED_SPEED', 'Snow_Priority', 'Number_Travel_Lanes', 
                      'Number_Park_Lanes', 'Number_Total_Lanes', 'TRUCK_ROUTE_TYPE']]
```

```python
p = pyproj.Proj('esri:102718')

def get_lon_lat(x1, x2, y1, y2):

    x = (x1 + x2)/2
    y = (y1 + y2)/2
    
    return p(x, y, inverse=True)
```

```python
geo_data2[['Lon', 'Lat']] = geo_data2.apply(lambda row : get_lon_lat(row['XFrom'], row['XTo'],
                                                                     row['YFrom'], row['YTo']), 
                                            axis = 1,
                                            result_type='expand')
```

```python
geo_data2 = geo_data2.drop(['XFrom', 'YFrom', 'XTo','YTo'], axis=1)
```

```python
geo_data2.head(15)
```

```python
geo_data2 = geo_data2.replace(' ', np.nan)
geo_data2 = geo_data2.replace('  ', '0')
```
